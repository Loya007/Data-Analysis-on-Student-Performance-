# Maths Score

```python
sns.distplot(df['math score'],label='Math Score')
```

```python
sns.boxplot(data = df, x = 'math score', orient = 'v')
```

```python
Q1, Q2, Q3, Q4 = np.percentile(df['math score'], np.arange(1,100,25))
Median = np.median(df['math score'])
Mean = np.mean(df['math score'])
```

# Observation

```python
print("The Quartile values are: \nQ1: {}\nQ2: {}\nQ3: {}\nQ4: {}".format(Q1, Q2, Q3, Q4))
print("\nThe Median Value is: ", Median)
print("The Mean Value is: ", Mean)
```

# Bivariate Analysis


Scores According To Gender

Reading Score and Writing Scores are highly correlated almost overlapping.

The data is not seperable linearly.

By observing the diagonal distribution plots we can assume that female students perform slightly better at reading and writing and male students perform a little better in math.





Scores According To Race/Ethnicity

```python
plt.figure(figsize=(7,7))
sns.pairplot(df , hue= 'race/ethnicity')
```

In the distribution plot of reading and writing scores we can observe that students from group D and E perform slightly better than other ethnicities.
